```python
#question1
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv ")
df1=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
```

```python
concat=pd.concat([df,df1],ignore_index=True)
concat.head(5)
```

```python
#question 2
concat=concat.dropna(subset=["Survived"])
concat
```

```python
#question 3
temp_df=concat["Cabin"].isnull()
concat["has_cabin"]=temp_df
for i in range (0,891):
    if concat["has_cabin"].iloc[i]:
        concat["has_cabin"].iloc[i]="NO"
    else:
        concat["has_cabin"].iloc[i]="YES"
```

```python
concat
```

```python
#question 4
import re
mr=r"Mr"
mrs=r"Mrs"
master=r"Master"
miss=r"Miss"
special=r"Special"
concat["Title"]=concat["Name"]
for i in range (0,891):
    match_mr=re.search(mr,concat["Name"].iloc[i])
    match_mrs=re.search(mrs,concat["Name"].iloc[i])
    match_master=re.search(master,concat["Name"].iloc[i])
    match_miss=re.search(miss,concat["Name"].iloc[i])
    match_special=re.search(special,concat["Name"].iloc[i])
    if match_mrs:
        concat["Title"].iloc[i]="Mrs"
    elif match_mr:
        concat["Title"].iloc[i]="Mr"
    elif match_master:
        concat["Title"].iloc[i]="Master"
    elif match_miss:
        concat["Title"].iloc[i]="Miss"
    elif match_special:
        concat["Title"].iloc[i]="Special"
```

```python
concat
```

```python
#Question 5
concat.fillna(0)
```

```python
#question 6
x=["below_14","below_19","below_22","below_25","below_28","below_31","below_36","below_42","below_50","below_80"]
concat["Categorical_age"]=pd.qcut(concat.Age,10,labels=x)
y=["below_7.6","below_7.9","below_8.1","below_10.6","below_14.5","below_21.7","below_27","below_41.6","below_78.1","below_512.4"]
concat["Categorical_fare"]=pd.qcut(concat.Fare,10,labels=y)
concat
```

```python
#question 7
dataset=concat.drop(["Cabin","Name","PassengerId", "Ticket", "Age", "Fare", "SibSp", "Parch"],1)
```

```python
dataset
```

```python
#question 8
dataset = pd.get_dummies( dataset)
dataset
```

```python
#question 9
import matplotlib.pyplot as plt
y= dataset.iloc[:,1]
x=dataset.drop("Survived",1)
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#question 10
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
pd.DataFrame({"Actual": list(y_test),"Predict":list(y_pred)})
```

```python
#question 11
classifier.score(x_test,y_test)
```
