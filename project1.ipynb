```python
import pandas as pd
import numpy as np
df=pd.read_csv("Big_mart.csv")
```

```python
df
```

```python
for i in range(0,8523):
    if df["Item_Fat_Content"].iloc[i]=="low fat":
        df["Item_Fat_Content"].iloc[i]="Low Fat"
    elif df["Item_Fat_Content"].iloc[i]=="LF":
        df["Item_Fat_Content"].iloc[i]="Low Fat"
    elif df["Item_Fat_Content"].iloc[i]=="reg":
        df["Item_Fat_Content"].iloc[i]="Regular"
```

```python
df
```

```python
dataset=df.dropna(subset=["Item_Weight"])
```

```python
x_train=dataset.iloc[:,[3,5]].values
y_train=dataset.iloc[:,1].values
```

```python
df_1=df.iloc[:,[1,3,5]]
df_1=df_1[pd.isnull(df_1).any(axis=1)]
```

```python
x_test=df_1.iloc[:,[1,2]].values
y_test=df_1.iloc[:,0].values
```

```python
pd.DataFrame(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
classifier = LinearRegression()
classifier.fit(x_train, y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred=y_pred.reshape(1463,1)
y_pred=pd.DataFrame(y_pred)
df.loc[df['Item_Weight'].isnull(),'Item_Weight']=y_pred[0]
df
```
